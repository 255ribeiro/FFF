```python
# entradas
# numero de vaos em x
vX = 300
# distancia dos vaos em x
dX = 5.25
# número de vãos em y
vY = 5
# distancia dos vaos em y
dY = 7.25
# numero pavimentos
vZ = 7
 # distancia piso a piso
dZ = 3.25
```

```python
# variáveis auxiliares
#nos da base em x e y
NxBase = vX +1
NyBase = vY +1

# nos nos pavimentos em x e y
Nx2 = vX*2 +1
Ny2 = vY*2 +1

#total de nos no modelo
Npts = (Nx2)*(Ny2)*(vZ) + ( NxBase * NyBase )
```

```python
mat = np.arange(0,Npts, 1, dtype=np.int32 )
```

```python
@nb.vectorize(["float32(int32, float32, float)"], target='parallel')
def coordX(M, NxB, NyB):
    if M <  (NxB*NyB):
        return  (M % NxB)*(dX)
    else:
        return  ((M - NxB*NyB) % Nx2)*(dX/2)
    
@nb.vectorize(["float32(int32)"], target='parallel')
def coordY(M):
    if M <  (NxBase*NyBase):
        return  ((M % (NxBase*NyBase))//NxBase)*(dY)
    else:
        return  (((M - NxBase*NyBase) % (Nx2*Ny2))//Nx2)*(dY/2)

@nb.vectorize(["float32(int32)"], target='parallel')
def coordZ(M):
    if M <  (NxBase*NyBase):
        return  0.0
    else:
        return  (((M - NxBase*NyBase) // (Nx2*Ny2)) + 1)*(dZ)
```

```python
cX = coordX(mat, NxBase)
cY = coordY(mat)
cZ = coordZ(mat)
print( '\ncoord X\n', cX , '\nCoord Y\n', cY, '\nCoord Z\n', cZ)
```
